```python
import numpy as np
from scipy.spatial import distance

pointA=np.array([2,4,6])
pointB=np.array([5,1,9])

euclidean=distance.euclidean(pointA,pointB)

print("Distance:",euclidean)
print("Similarity:",1/(1+euclidean))
```

```python
import numpy as np
from scipy.spatial import distance

pointA=np.array([2,4,6])
pointB=np.array([5,1,9])

manhattan=distance.cityblock(pointA,pointB)

print("Distance:",manhattan)
print("Similarity:",1/(1+manhattan))
```

```python
import numpy as np
from scipy.spatial import distance

pointA=np.array([2,4,6])
pointB=np.array([5,1,9])

minkowski=distance.minkowski(pointA,pointB,p=3)

print("Distance:",minkowski)
print("Similarity:",1/(1+minkowski))
```

```python
import pandas as pd
from scipy.stats import pearsonr

df=pd.DataFrame({
    'X':[10,20,30,40,50],
    'Y':[12,18,33,47,55]
})

corr,p=pearsonr(df['X'],df['Y'])

print(corr)
print(p)
```
