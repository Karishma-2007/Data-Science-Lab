```python
from sklearn.datasets import load_iris
import pandas as pd

iris=load_iris()

df=pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print(df.mean())
print(df.median())
print(df.mode().iloc[0])
```

```python
print(df.max()-df.min())
print(df.var())
print(df.std())
```

```python
print(df.skew())
print(df.kurt())
```
